```python
import pandas as pd
```

```python
df = pd.read_csv('churn-bigml-20.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Churn'] = df['Churn'].astype(int)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='International plan', y='Churn', data=df, ci=None)
plt.title('Churn Rate by International Plan')
plt.ylabel('Churn Rate (%)')
plt.xlabel('International Plan')

# Add percentage labels
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 0.01, f'{height:.2%}', ha='center')

plt.tight_layout()
plt.savefig('churn_by_intl_plan.png')
print("Generated churn_by_intl_plan.png")
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Voice mail plan', y='Churn', data=df, ci=None)
plt.title('Churn Rate by Voice Mail Plan')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Voice Mail Plan')

# Add percentage labels
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 0.01, f'{height:.2%}', ha='center')

plt.tight_layout()
plt.savefig('churn_by_voicemail_plan.png')
print("Generated churn_by_voicemail_plan.png")
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Churn', data=df, palette="viridis")
plt.title("Overall Churn Distribution")
plt.show()
```

```python


# Set a well-designed style and color palette for the plots
sns.set_style('whitegrid')
sns.set_palette('viridis')

# Convert 'Churn' from boolean to a more readable string
df['Churn'] = df['Churn'].replace({False: 'No Churn', True: 'Churn'})

# Create subplots for better organization
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Box plot for Churn vs. 'Total day charge'
sns.boxplot(x='Churn', y='Total day charge', data=df, ax=axes[0])
axes[0].set_title('Churn vs. Total Day Charge')
axes[0].set_ylabel('Total Day Charge')
axes[0].set_xlabel('Churn Status')

# Box plot for Churn vs. 'Customer service calls'
sns.boxplot(x='Churn', y='Customer service calls', data=df, ax=axes[1])
axes[1].set_title('Churn vs. Customer Service Calls')
axes[1].set_ylabel('Number of Customer Service Calls')
axes[1].set_xlabel('Churn Status')

plt.tight_layout()
plt.savefig('styled_churn_analysis.png')
print("Generated styled_churn_analysis.png")
```

```python


# Set a style for the plots
sns.set_style('whitegrid')

# Convert 'Churn' from boolean to a more readable string
df['Churn'] = df['Churn'].replace({False: 'No Churn', True: 'Churn'})

# Create subplots for better organization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histogram for Total day minutes, separated by Churn status
sns.histplot(data=df, x='Total day minutes', hue='Churn', kde=True, ax=axes[0])
axes[0].set_title('Distribution of Total Day Minutes by Churn Status')
axes[0].set_xlabel('Total Day Minutes')
axes[0].set_ylabel('Count')

# Histogram for Total day charge, separated by Churn status
sns.histplot(data=df, x='Total day charge', hue='Churn', kde=True, ax=axes[1])
axes[1].set_title('Distribution of Total Day Charge by Churn Status')
axes[1].set_xlabel('Total Day Charge')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('churn_histograms.png')
print("Generated churn_histograms.png")
```

```python


# Convert 'Churn' from boolean to integer (1 for True, 0 for False) for analysis
df['Churn'] = df['Churn'].astype(int)

# Calculate churn rate by customer service calls, then sort the results
churn_by_calls = df.groupby('Customer service calls')['Churn'].mean().reset_index()
churn_by_calls = churn_by_calls.sort_values('Customer service calls')

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Customer service calls', y='Churn', data=churn_by_calls)
plt.title('Churn Rate by Number of Customer Service Calls')
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Churn Rate (%)')

# Add percentage labels
for index, row in churn_by_calls.iterrows():
    plt.text(index, row['Churn'] + 0.01, f'{row["Churn"]:.2%}', ha='center')

plt.tight_layout()
plt.savefig('churn_by_service_calls.png')
print("Generated churn_by_service_calls.png")
```

```python




# Convert boolean columns to a more descriptive string for plotting
df['International plan'] = df['International plan'].replace({False: 'No', True: 'Yes'})
df['Voice mail plan'] = df['Voice mail plan'].replace({False: 'No', True: 'Yes'})
df['Churn'] = df['Churn'].replace({False: 'No Churn', True: 'Churn'})

# Create a cross-tabulation of churn status by international and voice mail plans
churn_plan_cross = pd.crosstab(
    [df['International plan'], df['Voice mail plan']],
    df['Churn']
)

# Plot the stacked bar chart
churn_plan_cross.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Churn Status by International and Voice Mail Plans')
plt.xlabel('International Plan, Voice Mail Plan')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.legend(title='Churn Status')
plt.tight_layout()
plt.savefig('stacked_churn_by_plan.png')
print("Generated stacked_churn_by_plan.png")
```

```python


# Convert 'Churn' from boolean to integer (1 for True, 0 for False) for analysis
df['Churn'] = df['Churn'].astype(int)

# Create a bar plot of churn rate by Area code
plt.figure(figsize=(8, 6))
sns.barplot(x='Area code', y='Churn', data=df, ci=None)
plt.title('Churn Rate by Area Code')
plt.xlabel('Area Code')
plt.ylabel('Churn Rate (%)')

# Add percentage labels
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 0.01, f'{height:.2%}', ha='center')

plt.tight_layout()
plt.savefig('churn_by_area_code.png')
print("Generated churn_by_area_code.png")
```
